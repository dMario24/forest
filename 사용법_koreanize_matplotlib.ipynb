```python
import matplotlib.pyplot as plt
import koreanize_matplotlib

# 산림과학 데이터 준비 (예시)
x = ['소나무', '참나무', '단풍나무', '잣나무'] # 나무 종류
y = [50, 75, 60, 45] # 나무 수 (단위: 1000그루)

# 그래프 생성
plt.figure(figsize=(8, 6))
plt.bar(x, y, color=['forestgreen', 'saddlebrown', 'darkred', 'olivedrab'])

# 제목 및 라벨 설정
plt.title('주요 나무 종류별 개체 수', fontsize=15) # 제목 변경
plt.xlabel('나무 종류', fontsize=12) # x축 라벨 변경
plt.ylabel('개체 수 (천 그루)', fontsize=12) # y축 라벨 변경

# 그리드 추가
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 그래프 표시
plt.show()
```
